```python
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from skimage.io import imread, imshow
from collections import Counter
%matplotlib inline
```

```python
with open('../data/furniture_train.json','rb') as f:
  json_trn = json.load(f)
  
with open('../data/furniture_validation.json','rb') as f:
  json_val = json.load(f)
```

```python
json_trn['annotations'][0]
```

```python
trn_labels = pd.DataFrame([int(i['label_id']) for i in json_trn['annotations']])
val_labels = pd.DataFrame([int(i['label_id']) for i in json_val['annotations']])
```

```python
trn_labels.hist(bins = 128)
plt.xlabel('Class')
plt.title('Train data')
plt.ylabel('Amount in Class')
plt.savefig('trainhisto.png')
```

```python
val_labels.hist(bins = 128)
plt.xlabel('Class')
plt.title('Test data')
plt.ylabel('Amount in Class')
plt.savefig('testhost.png')
```
